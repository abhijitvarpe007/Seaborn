### Seaborn Barplot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
sns.barplot()
plt.show()
```

```python
sns.barplot(x=df.total_bill)
plt.show()
```

```python
sns.barplot(x=df.day,y=df.total_bill)
plt.show()
```

```python
sns.barplot(x=df.day,y=df.total_bill,hue=df.sex)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue=df.sex,data=df)
plt.show()
```

```python
order=['Sun','Sat','Fri','Thur']
sns.barplot(x='day',y='total_bill',hue=df.sex,order=order,data=df)
plt.show()
```

```python
hue_order=['Female','Male']
sns.barplot(x='day',y='total_bill',hue=df.sex,order=order,
            hue_order=hue_order,data=df)
plt.show()
```

```python
import numpy as np

sns.barplot(x='day',y='total_bill',hue=df.sex,order=order,data=df,estimator=np.mean)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue=df.sex,order=order,data=df,estimator=np.max)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue=df.sex,
            ci=14,data=df)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue=df.sex,
            ci=14,n_boot=2,data=df)
plt.show()
```

```python
sns.barplot(y='day',x='total_bill',hue=df.sex,
             data=df)
plt.show()
```

```python
sns.barplot(x='total_bill',y='size',hue='sex',data=df
           ,orient='h')
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue=df.sex,
             palette='hot',data=df)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue=df.sex,
             palette='magma',data=df)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue=df.sex,
             saturation=5,data=df)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue=df.sex,
             color='m',errcolor='0.5',data=df)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue=df.sex,
             color='m',errwidth='5',data=df)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue=df.sex,
             color='m',capsize=1,data=df)
plt.show()
```

```python
sns.barplot(x='day',y='total_bill',hue=df.sex,
             color='m',dodge=False,data=df)
plt.show()
```

```python
kwargs={'alpha':0.9,'linestyle':':','linewidth':2,'edgecolor':'k'}
sns.barplot(x='day',y='total_bill',
           data=df,**kwargs)
plt.show()
```

```python
kwargs={'alpha':1,'linestyle':':','linewidth':2,'edgecolor':'k'}
ax=sns.barplot(x='day',y='total_bill',
             data=df,**kwargs)
ax.set(title='Barplot of tips DataFrame',
      xlabel='Days',ylabel='Total_bill')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.barplot(x='day',y='total_bill',
             data=df,**kwargs)
plt.title('Barplot of tips DataFrame',fontsize=20)
plt.xlabel("Days",fontsize=15)
plt.ylabel('Total_Bill',fontsize=15)
plt.show()
```
