### Seaborn Histogram / Displot

```python
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
df=sns.load_dataset('tips')
df
```

```python
sns.distplot(df['size'])
plt.show()
```

```python
sns.distplot(df['total_bill'])
plt.show()
```

```python
sns.distplot(df['total_bill'],bins=10)
plt.show()
```

```python
sns.distplot(df['total_bill'],bins=100)
plt.show()
```

```python
sns.distplot(df['total_bill'],hist=False)
plt.show()
```

```python
sns.distplot(df['total_bill'],kde=False)
plt.show()
```

```python
sns.distplot(df['total_bill'],rug=True)
plt.show()
```

```python
sns.distplot(df['total_bill'],fit=norm,kde=False)
plt.show()
```

```python
sns.distplot(df['total_bill'],color='r')
plt.show()
```

```python
sns.distplot(df['total_bill'],color='r',vertical=True)
plt.show()
```

```python
sns.distplot(df['total_bill'],color='r',norm_hist=True)
plt.show()
```

```python
sns.distplot(df['total_bill'],color='m',axlabel='Total Bill')
plt.show()
```

```python
sns.distplot(df['total_bill'],color='m',label='Total Bill')
plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.set()
sns.distplot(df['total_bill'],label='Total Bill')
plt.title("Histogram of Total Bill")
plt.legend()
plt.show()
```

```python
df.total_bill.sort_values()
```

```python
bins=[1,5,10,15,20,25,30,35,40,45,50,55]
plt.figure(figsize=(12,7))
sns.set()
sns.distplot(df['total_bill'],label='Total Bill',bins=bins)
plt.title("Histogram of Total Bill")
plt.xticks(bins)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.set()
sns.distplot(df['total_bill'],
            hist_kws={'color':'#DC143C','edgecolor':'#aaff00','linewidth':5,'linestyle':'--','alpha':0.8})
plt.title("Histogram of Total Bill")
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.set()
sns.distplot(df['total_bill'],
            hist_kws={'color':'#DC143C','edgecolor':'#aaff00','linewidth':5,'linestyle':'--','alpha':0.8},
            kde_kws={'color':'#8e00ce','linewidth':4,'linestyle':'--','alpha':0.8})
plt.title("Histogram of Total Bill")
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.set()
sns.distplot(df['total_bill'],
            hist_kws={'color':'#DC143C','edgecolor':'#aaff00','linewidth':5,'linestyle':'--','alpha':0.8},
            kde_kws={'color':'#8e00ce','linewidth':4,'linestyle':'--','alpha':0.8},
            rug=True,
            rug_kws={'color':'#0426d0','edgecolor':'#00dbff','linewidth':5,'linestyle':'--','alpha':0.8})
plt.title("Histogram of Total Bill")
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.set()
sns.distplot(df['total_bill'],
            hist_kws={'color':'#DC143C','edgecolor':'#aaff00','linewidth':5,'linestyle':'--','alpha':0.8},
            kde=False,
            fit=norm,
            fit_kws={'color':'#8e00ce','linewidth':4,'linestyle':'--','alpha':0.8},
            
            rug=True,
            rug_kws={'color':'#0426d0','edgecolor':'#00dbff','linewidth':5,'linestyle':'--','alpha':0.8})
plt.title("Histogram of Total Bill")
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.set()
sns.distplot(df['total_bill'],
            hist_kws={'color':'#DC143C','edgecolor':'#aaff00','linewidth':5,'linestyle':'--','alpha':0.8},
            kde=False,
            fit=norm,
            fit_kws={'color':'#8e00ce','linewidth':4,'linestyle':'--','alpha':0.8},
            
            rug=True,
            rug_kws={'color':'#0426d0','edgecolor':'#00dbff','linewidth':5,'linestyle':'--','alpha':0.8},
            label='TB')
plt.title("Histogram of Total Bill",fontsize=15)
plt.xlabel("Total Bill")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.distplot(df['total_bill'],label='Total Bill',bins=bins)
sns.distplot(df['tip'],label='Tips')
sns.distplot(df['size'],label='Size')
plt.legend()
plt.show()
```
