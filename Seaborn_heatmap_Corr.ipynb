### Correlation
A correlation is a statistical measure of the relationship between two variables(X,Y)

The Correation coefficient is value from -1 to 1

-1: prefect negative correlation (X:increases then Y-decreases)

0:No correlation (X:increases no affect on Y and Vice versa)

1:prefect positive correaltion(X: increases then Y increases)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Warming1.csv")
df.head()
```

```python
df.corr().head()
```

```python
arr_2d=np.linspace(1,5,20).reshape(4,5)
arr_2d
```

```python
plt.figure(figsize=(12,7))
ax=sns.heatmap(arr_2d,annot=True,linewidth=3)
ax.tick_params(size=10,color='k',labelsize=10,labelcolor='k')
plt.show()
```

```python
from sklearn.datasets import load_breast_cancer
cancer_dataset=load_breast_cancer()
cancer_dataset
```

```python
cancer_df=pd.DataFrame(np.c_[cancer_dataset['data'],cancer_dataset['target']],
                      columns=np.append(cancer_dataset['feature_names'],['target']))
cancer_df
```

```python
plt.figure(figsize=(18,10))
ax=sns.heatmap(cancer_df.corr(),annot=True,linewidth=1)
ax.tick_params(size=10,color='m',labelsize=10,labelcolor='k')
plt.title("Heatmap of Breast Cancer Patients",fontsize=20)
plt.show()
```
