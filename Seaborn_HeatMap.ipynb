### Seabor Heatmap
Heatmap is a graphical represenation of 2D Data. Each data values represent in a matrix and its has a special color

```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr_2d=np.linspace(1,5,12).reshape(4,3)
arr_2d
```

```python
sns.heatmap(arr_2d,annot=True)
plt.show()
```

```python
global_warming=pd.read_csv('Warming1.csv')
global_warming.head()
```

```python
global_warming.dtypes
```

```python
df=global_warming.drop(columns=['Country Code','Indicator Name','Indicator Code'],axis=1).set_index("Country Name")
df
```

```python
plt.figure(figsize=(14,9))
sns.heatmap(df)
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(arr_2d,annot=True,vmin=0,vmax=10)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(arr_2d,annot=True,cmap='coolwarm')
plt.show()
```

```python
arr_2d
```

```python
annot_arr=np.array([['a01','a02','a03'],['a11','a12','a13'],['a21','a22','a23'],['a31','a32','a33']])
annot_arr
```

```python
sns.heatmap(arr_2d,annot=annot_arr,fmt='5')
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.heatmap(arr_2d,annot=annot_arr,fmt='5')
plt.show()
```

```python
plt.figure(figsize=(12,7))
annot_kws={'fontsize':10,'fontstyle':'italic',
          'color':'k','alpha':0.9,
          'verticalalignment':'center','backgroundcolor':'w'}
sns.heatmap(arr_2d, cmap='coolwarm',annot=True,annot_kws=annot_kws)
plt.show()
```

```python
# adding linesapce and linewidth
plt.figure(figsize=(12,7))
sns.heatmap(arr_2d,annot=True,cmap='coolwarm',linewidth=4,linecolor='c')
plt.show()
```

```python
# hide the color bar
plt.figure(figsize=(12,7))
sns.heatmap(arr_2d,annot=True,cmap='coolwarm',linewidth=4,linecolor='c',cbar=False)
plt.show()
```

```python
# Hiding the X and Y axis notation
plt.figure(figsize=(12,7))
sns.heatmap(arr_2d,annot=True,cmap='coolwarm',cbar=False,xticklabels=False,yticklabels=False)
plt.show()
```

```python
plt.figure(figsize=(12,7))
cbar_kws={"orientation":'horizontal','shrink':1,
         'extend':'min','extendfrac':0.1,'ticks':np.arange(0,22),
         'drawedges':True}
sns.heatmap(arr_2d,annot=True,cbar_kws=cbar_kws)
plt.show()
```

```python

plt.figure(figsize=(12,7))
ax=sns.heatmap(arr_2d,annot=True,)
ax.set(title="HeatMap",
      xlabel='Values',
      ylabel='Values')
sns.set(font_scale=1)
plt.show()
```
