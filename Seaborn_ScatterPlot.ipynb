### Seaborn Scatter Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('titanic')
df
```

```python
sns.scatterplot(x='age',y='fare',color='r',data=df)
plt.show()
```

```python
sns.scatterplot(x='age',y='fare',hue='sex',data=df)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(x='age',y='fare',hue='sex',
                style='who',size='who',data=df)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(x='age',y='fare',hue='sex',
                style='who',size='who',sizes=(100,200),data=df)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(x='who',y='fare',hue='alive',
                style='alive',size='who',sizes=(100,200),data=df)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(x='who',y='fare',hue='alive',
                style='alive',size='who',sizes=(100,200),
                palette='inferno_r',alpha=0.7,data=df)
plt.show()
```

```python
plt.figure(figsize=(12,7))
sns.scatterplot(x='age',y='fare',data=df)
sns.lineplot(x='age',y='fare',data=df)
sns.barplot(x='age',y='fare',data=df)
plt.show()
```
